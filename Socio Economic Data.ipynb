```python
import pandas as pd
import matplotlib.pyplot as plt

happy_df = pd.read_csv("happiness_df.csv")

def get_happiness_scores(country_name):
    mask = happy_df['Country'] == country_name
    happiness_scores = happy_df.loc[mask, 'Happiness Score'].tolist()
    return happiness_scores

def plot_country(happiness_scores):
    years = range(2018, 2023)
    plt.plot(years, happiness_scores)
    plt.xlabel('Year')
    plt.ylabel('Happiness Scores')
    plt.title('Change in Happiness from 2015 to 2022')
    plt.show()

happy_df = happy_df[happy_df["Year"] >= 2018]
happy_df
```

```python

import numpy as np

#internet Usage
broadband_df = pd.read_csv("broadband_per_100.csv")

#health
health_df = pd.read_csv("physicians_per_1000.csv")
health_df
```

```python
happiness_df = pd.read_csv("happiness_df.csv")
```

```python
broadband_df = broadband_df[broadband_df["TIME"].str.endswith('Q2')]
broadband_df
```

```python
not_included = []
for country in broadband_df["Country"].unique():
    if country not in happiness_df["Country"].unique():
        not_included.append(country)
for country in happiness_df["Country"].unique():
    if country not in broadband_df["Country"].unique():
        not_included.append(country)
print(not_included)
```

```python


# Define the years of interest
years = ['2018', '2019', '2020', '2021', '2022']

# Initialize an empty dictionary to hold the data
data = {'Country': [],'2018':[], '2019': [], '2020': [], '2021': [], '2022': []}

# Loop over each unique country in broadband_df
for country in broadband_df['Country'].unique():

    # Add the country name to the dictionary
    data['Country'].append(country)

    # Loop over each year of interest
    for year in years:

        # Extract the value for the given country and year
        value = broadband_df[(broadband_df['Country'] == country) & (broadband_df['TIME'].str.startswith(year))]['Value'].iloc[0]

        # Add the value to the dictionary
        data[year].append(value)

# Create a new dataframe from the dictionary
broadband_df = pd.DataFrame(data)

broadband_df
```

```python
# Sort the new dataframe by the mean value across all years
sorted_df = broadband_df.copy()
sorted_df = sorted_df.sort_values("2022")
# Select the top and bottom 10 countries
bottom_10 = sorted_df.head(10)
top_10 = sorted_df.tail(10)

# Create a bar plot of the top 10 countries
ax1 = top_10.plot(x='Country', y=years, kind='bar')
ax1.set_title('Top 10 Countries')
ax1.set_xlabel('Country')
ax1.set_ylabel('Broadband Value')

# Create a bar plot of the bottom 10 countries
ax2 = bottom_10.plot(x='Country', y=years, kind='bar')
ax2.set_title('Bottom 10 Countries')
ax2.set_xlabel('Country')
ax2.set_ylabel('Broadband Value')

# Display the plots
plt.show()
```

```python
broadband_df['Increase'] = broadband_df['2022'] - broadband_df['2018']

# Sort the new dataframe by the increase in broadband value from 2019 to 2022
sorted_df = broadband_df.sort_values('Increase', ascending=False)

# Select the top 10 countries with the largest increase in broadband value
top_10_increase = sorted_df.head(10)

# Create a bar plot of the top 10 countries with the largest increase in broadband value
ax3 = top_10_increase.plot(x='Country', y="Increase", kind='bar')
ax3.set_title('Increase in Broadband per 100 people from 2018 to 2022')
ax3.set_xlabel('Country')
ax3.set_ylabel('Change in Broadband Value')

# Display the plots
plt.show()
broadband_inc_arr = list(["Czech Republic", "Portugal", "Slovak Republic", "Turkiye", "Greece", "Estonia", "Chile", "Mexico", "United States", "Hungary"])

```

```python
# Get a list of unique countries in the original data frame
countries = happy_df['Country'].unique()

# Create a new data frame with columns for each year
new_df = pd.DataFrame(columns=['Country', '2018', '2019', '2020', '2021', '2022'])

# Iterate over each country and year, and fill in the values in the new data frame
for country in countries:
    country_values = [country]
    for year in range(2018, 2023):
        value = happy_df[(happy_df['Country'] == country) & (happy_df['Year'] == year)]['Happiness Score']
        if len(value) > 0:
            country_values.append(value.values[0])
        else:
            country_values.append(None)
    new_df.loc[len(new_df)] = country_values

top_10_increase_happiness = new_df[new_df['Country'].isin(top_10_increase['Country'])]

top_10_increase_happiness.set_index('Country', inplace=True)

# Create a line plot for each country, showing the happiness score over time
ax5 = top_10_increase_happiness.T.plot(kind='line')

# Set the title, x-axis label, and y-axis label for the plot
ax5.set_title('Happiness Score over Time for Top 10 Countries with Largest Increase in Broadband Value')
ax5.set_xlabel('Year')
ax5.set_ylabel('Happiness Score')

# Display the plot
plt.show()
```

Increase in broadband access per 1000 people has no correlation in happiness in the range of years from 2018-2022. This data looks at the countries with the highest broadband score and charts their happiness. I hypothesized when I first looked at this data that the countries with the largest increase in broadband would see sizable increases in their happiness. However, looking into the broadband table, a large majority of the countries that appear in the happiness table do not have their broadband statistics listed. Turkey, Slovak Republic, and Czechia, three countries in the top 10 increase in broadband value, also do not appear in the happiness dataset in the range of years mentioned. Covid-19 also played a major factor in data being more difficult to analyze in its relationship with happiness because every country was going through drastic changes from 2020 onwards, and it was harder for data to be collected accurately, and less countries reported their data.

More observations made about broadband data is that a lot of the countries in the top 10 of most improved for broadband statistics, such as Mexico, still appear in the bottom 10 countries for broadband access in 2022 of the countries included in the data.
