```python

import numpy as np

#internet Usage
broadband_df = pd.read_csv("broadband_per_100.csv")

#health
health_df = pd.read_csv("physicians_per_1000.csv")
health_df
```

```python
happiness_df = pd.read_csv("happiness_df.csv")
```

```python
broadband_df = broadband_df[broadband_df["TIME"].str.endswith('Q2')]
broadband_df
```

```python
not_included = []
for country in broadband_df["Country"].unique():
    if country not in happiness_df["Country"].unique():
        not_included.append(country)
for country in happiness_df["Country"].unique():
    if country not in broadband_df["Country"].unique():
        not_included.append(country)
print(not_included)
```

```python


# Define the years of interest
years = ['2018', '2019', '2020', '2021', '2022']

# Initialize an empty dictionary to hold the data
data = {'Country': [],'2018':[], '2019': [], '2020': [], '2021': [], '2022': []}

# Loop over each unique country in broadband_df
for country in broadband_df['Country'].unique():

    # Add the country name to the dictionary
    data['Country'].append(country)

    # Loop over each year of interest
    for year in years:

        # Extract the value for the given country and year
        value = broadband_df[(broadband_df['Country'] == country) & (broadband_df['TIME'].str.startswith(year))]['Value'].iloc[0]

        # Add the value to the dictionary
        data[year].append(value)

# Create a new dataframe from the dictionary
broadband_df = pd.DataFrame(data)

broadband_df
```

```python
merged_data = pd.merge(broadband_df, happy_df, on="Country")
merged_data = merged_data[merged_data["Year"] == 2022]
merged_data = merged_data.drop(["Year", "Unnamed: 0"], axis=1)
merged_data
```

```python
import numpy as np
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

# Create a scatter plot
plt.scatter(merged_data["Happiness Score"], merged_data['2022'])

for i, row in merged_data.iterrows():
    country = row['Country']
    plt.text(row["Happiness Score"], row['2022'], country)

# Calculate the line of best fit
slope, intercept = np.polyfit(merged_data["Happiness Score"], merged_data['2022'], 1)
line_of_best_fit = slope * merged_data["Happiness Score"] + intercept

# Plot the line of best fit
plt.plot(merged_data["Happiness Score"], line_of_best_fit, color='red')

# Set axis labels
plt.xlabel('Happiness Score')
plt.ylabel('Broadband Access per 100 people in 2022')

# Set the plot title
plt.title('Correlation Between Broadband Access and Happiness in 2022')

# Calculate the correlation coefficient
correlation, p_value = pearsonr(merged_data["Happiness Score"], merged_data['2022'])
print("Correlation coefficient:", correlation)
print("p-value: ", p_value)
# Display the plot
plt.show()
```

The correlation between countries with the highest rate of broadband access and the highest happiness score has a pearsonr correlation coefficient of 0.498 which suggests that there is a moderate correlation between broadband access in 2022 and happiness in 2022. The P-value for the correlation coefficient with 32 total datapoints is: 0.004. This P-Value is less than the accepted standard of 0.05, therefore implies that the correlation between these two variables was not merely decided by random chance.
