```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
happy_df = pd.read_csv("happiness_df.csv")

def get_happiness_scores(country_name):
    mask = happy_df['Country'] == country_name
    happiness_scores = happy_df.loc[mask, 'Happiness Score'].tolist()
    return happiness_scores

def plot_country(happiness_scores):
    years = range(2015, 2023)
    plt.plot(years, happiness_scores)
    plt.xlabel('Year')
    plt.ylabel('Happiness Scores')
    plt.title('Change in Happiness from 2015 to 2022')
    plt.show()

# Get a list of unique countries in the original data frame
countries = happy_df['Country'].unique()

# Create a new data frame with columns for each year
new_df = pd.DataFrame(columns=['Country', "2015", "2016", "2017", "2018", "2019", "2020", "2021", "2022"])

for country in countries:
    country_values = [country]
    for year in range(2015, 2023):
        value = happy_df[(happy_df['Country'] == country) & (happy_df['Year'] == year)]['Happiness Score']
        if len(value) > 0:
            country_values.append(value.values[0])
        else:
            country_values.append(None)
    new_df.loc[len(new_df)] = country_values
```

```python
work_life = pd.read_csv("work-life.csv")

work_life = work_life[work_life["ISC11A"] == "T"]
work_life = work_life[work_life["MEASURE"] == "VALUE"]
family_resp = work_life[work_life["INDICATOR"] == "EQLS_PERC_NEG_FAMILY"]
work_conc = work_life[work_life["INDICATOR"] == "EQLS_PERC_NEG_JOB"]
work_life
```

```python
family_resp = family_resp.sort_values("Value")
family_resp
```

```python
work_conc = work_conc.sort_values("Value")
work_conc
```

```python
import pandas as pd

# Assuming work_conc and family_resp are the existing DataFrames

# Merge the DataFrames on "Country" column
merged_df = pd.merge(work_conc, family_resp, on="Country")
merged_df = pd.merge(merged_df, new_df, on="Country")
# Create a new DataFrame with desired columns
work_life2 = pd.DataFrame()
work_life2["Country"] = merged_df["Country"]
work_life2["work_conc"] = merged_df["Value_x"]
work_life2["family_resp"] = merged_df["Value_y"]
work_life2["Average"] = (merged_df["Value_x"] + merged_df["Value_y"]) / 2
work_life2["Happiness"] = merged_df["2015"]

# Print the new DataFrame
work_life2.sort_values("Average", ascending=False)
```

```python
import numpy as np
from scipy.stats import pearsonr

# Assuming work_life2 is the DataFrame containing "Average", "Happiness", and "Country" columns

# Create a scatter plot
plt.scatter(work_life2['Happiness'], work_life2['Average'])

# Add labels for each point
for i, country in enumerate(work_life2['Country']):
    plt.text(work_life2['Happiness'][i], work_life2['Average'][i], country)

# Calculate the line of best fit
slope, intercept = np.polyfit(work_life2['Happiness'], work_life2['Average'], 1)
line_of_best_fit = slope * work_life2['Happiness'] + intercept

# Plot the line of best fit
plt.plot(work_life2['Happiness'], line_of_best_fit, color='red')

# Set axis labels
plt.xlabel('Happiness')
plt.ylabel('% Average Unsatisfied')

# Set the plot title
plt.title('Correlation Between Work-Life and Happiness')

# Calculate the correlation coefficient
correlation, p_value = pearsonr(work_life2['Happiness'], work_life2['Average'])
print("Correlation coefficient:", correlation)
print("p-value: ", p_value)
# Display the plot
plt.show()
```

The pearson(r) correlation coefficient is -0.548
The P-Value is .005. The result is significant at p < .05.

This graph shows the Average between percent of people who feel like work is interfering with their personal life, and the percent of people who think their personal life is interfering with their work. That average is plotted against each country's happiness value for the year 2015, which is the year the work-life study was completed. The result is a moderate, yet significant negative correlation between the two variables.
